```python
if 'gender' in data.columns and 'age' in data.columns:
    # Agrupar los datos por 'age' y encontrar el valor más frecuente de 'gender' para cada grupo de 'age'
    most_frequent_gender = data.groupby('age')['gender'].agg(lambda x: x.mode()[0] if not x.mode().empty else None)

    # Función para completar los valores nulos en 'gender'
    def fill_gender(row):
        if pd.isnull(row['gender']):
            return most_frequent_gender[row['age']]
        else:
            return row['gender']

    # Aplicar la función para llenar los valores nulos en 'gender'
    data['gender'] = data.apply(fill_gender, axis=1)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Leer el archivo CSV desde la URL
url = "https://raw.githubusercontent.com/RosaGaLl/DataScience/main/churn_prediction.csv"
data = pd.read_csv(url)

# Verificar los datos
print(data.head())

# Gráfico de dispersión para visualizar la relación entre edad, género y tasa de abandono
plt.figure(figsize=(12, 6))
sns.scatterplot(x='age', y='churn', hue='gender', data=data)
plt.title('Relación entre Edad, Género y Tasa de Abandono')
plt.xlabel('Edad')
plt.ylabel('Tasa de Abandono')
plt.legend(title='Género')
plt.show()
```
