**DATA DE ABANDONO DE CLIENTES BANCARIOS**

## Descripción del Dataset

Este dataset contiene información sobre aproximadamente 28,382 clientes de una empresa financiera. Conteniendo datos significativos como el tiempo de estadía, edad, ocupación y otra información crucial del cliente para una análisis predictivo con respecto al abandono luego de la apertura su cuenta. En base a ello:
**El objetivo es analizar el comportamiento de los clientes y determinar factores que influyen en el abandono del servicio (churn), con el fin de tomar medidas correctivas al igual que evitar que una gran cantidad de clientes opten por dejar de usar los servicios del banco.**

### Variables del Dataset

- **customer_id**: Identificador único para cada cliente.
- **vintage**: Duración de la relación del cliente con la empresa.
- **edad**: Edad del cliente.
- **género**: Género del cliente.
- **dependientes**: Número de dependientes que tiene el cliente.
- **ocupación**: Ocupación del cliente.
- **ciudad**: Ciudad en la que se encuentra el cliente.
- **customer_nw_category**: Categoría de patrimonio neto del cliente.
- **branch_code**: Código que identifica la sucursal asociada al cliente.
- **current_balance**: Saldo actual en la cuenta del cliente.
- **previous_month_end_balance**: Saldo de la cuenta al final del mes anterior.
- **Average_monthly_balance_prevQ**: Saldo mensual promedio en el trimestre anterior.
- **promedio_saldo_mensual_prevQ2**: Saldo promedio mensual en el segundo trimestre anterior.
- **current_month_credit**: Monto del crédito en el mes actual.
- **previous_month_credit**: Monto del crédito en el mes anterior.
- **current_month_debit**: Importe del débito en el mes actual.
- **previous_month_debit**: Importe del débito del mes anterior.
- **current_month_balance**: Saldo de la cuenta en el mes actual.
- **previous_month_balance**: Saldo de la cuenta en el mes anterior.
- **churn**: La variable de destino que indica si el cliente ha abandonado el servicio (1 para abandonado, 0 para no abandonado).
- **last_transaction**: Marca de tiempo de la última transacción del cliente.


```python
# Inserción de librerías pandas , seaborn y matplotlib, al igual que la muestra de la base de datos.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/RosaGaLl/DataScience/main/churn_prediction.csv"
data = pd.read_csv(url)
data

```

```python
# Descripción estadística del dataset
data.describe()

# Información sobre las variables del dataset
data.info()
```

```python
#Se cambia la columna fecha last_transaction para que tenga el tipo de dato correcto
data['last_transaction']=pd.to_datetime(data['last_transaction'])
```

```python
Datos_nan = data.isnull().sum()
print("\nValores vacíos por columna:")
Datos_nan
```

```python
#Ante valores NAN en el dataset, se empezará a manejaros llenándolos con un lógica como:
#gender 27857
data.info()
```

```python
if 'gender' in data.columns and 'age' in data.columns:
    # Se agrupan los datos por 'age' y se encuentra el valor más frecuente de 'gender' para cada grupo de 'age'
    most_frequent_gender = data.groupby('age')['gender'].agg(lambda x: x.mode()[0] if not x.mode().empty else None)

    # Se da una función para completar los valores nulos en 'gender'
    def fill_gender(row):
        if pd.isnull(row['gender']):
            return most_frequent_gender[row['age']]
        else:
            return row['gender']

    # Posteriormente, se aplica la función para llenar los valores nulos en 'gender'
    data['gender'] = data.apply(fill_gender, axis=1)
```

```python
# Para la columna DEPENDENTS
average_dependents_by_nw_Category=data.groupby('customer_nw_category')['dependents'].max()
def fill_dependents(row):
    if pd.isnull(row['dependents']) and row['customer_nw_category'] in average_dependents_by_nw_Category:
        return average_dependents_by_nw_Category[row['customer_nw_category']]
    else:
        return row['dependents']

data['dependents'] = data.apply(fill_dependents, axis=1)
```

```python
# Para la columna OCCUPATION
moda_occupation_by_gender=data.groupby('gender')['occupation'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)


def fill_occupation(row):
    if pd.isnull(row['occupation']) :
       return moda_occupation_by_gender.get(row['gender'], row['occupation'])
    else:
        return row['occupation']
data['occupation'] = data.apply(fill_occupation, axis=1)
```

```python
# Para la columna CITY
#Se dará de acuerdo a la moda de la ciudad más frecuente por occupation
moda_city_by_occupation = data.groupby('occupation')['city'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
def fill_city(row):
    if pd.isnull(row['city']):
        return moda_city_by_occupation.get(row['occupation'], row['city'])
    else:
        return row['city']

data['city'] = data.apply(fill_city, axis=1)
```

**Rad Chart de variables edad, saldo actual y saldo promedio mensual del trimestre anterior por churn**

```python
import plotly.graph_objects as go

# Normalizar los datos para que sean comparables en el radar chart
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

# Seleccionar las variables de interés
variables = ['age', 'current_balance', 'average_monthly_balance_prevQ']
data_normalized = data.copy()
for var in variables:
    data_normalized[var] = normalize(data[var])

# Agrupar por churn y calcular las medias
data_grouped = data_normalized.groupby('churn')[variables].mean().reset_index()

# Radar chart
fig = go.Figure()

for i in range(len(data_grouped)):
    fig.add_trace(go.Scatterpolar(
        r=data_grouped.loc[i, variables].values,
        theta=variables,
        fill='toself',
        name=f'Churn: {data_grouped["churn"].iloc[i]}'
    ))

fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 1]
        )),
    showlegend=True,
    title="Radar Chart de Variables por Churn"
)

fig.show()
```

El radar chart proporciona una vista de conjunto de cómo interactúan las variables edad, current_balance y average_monthly_balance_prevQ para los clientes que abandonan y los que no. Al visualizar estas variables juntas, podemos identificar patrones complejos y ver si, por ejemplo, los clientes jóvenes con bajos saldos actuales y bajos saldos promedio mensuales del trimestre anterior tienen más probabilidades de abandonar.

```python
#HIPOTESIS 1
from scipy.stats import ttest_ind

# Dividir los datos en dos grupos: clientes que han abandonado y los que no
churn_yes = data[data['churn'] == 1]['age']
churn_no = data[data['churn'] == 0]['age']

# Prueba t para comparar las edades de ambos grupos
t_stat, p_val = ttest_ind(churn_yes, churn_no)
print(f'T-statistic: {t_stat}, P-value: {p_val}')
```

Como p-value de la prueba t es menor a 0.05, podemos concluir que hay una diferencia estadísticamente significativa entre las edades de los clientes que abandonaron y los que no, lo que apoya nuestra hipótesis de que la edad es una variable relevante.



```python
#HIPOTESIS 2
# Dividir los datos en dos grupos: clientes que han abandonado y los que no
balance_churn_yes = data[data['churn'] == 1]['average_monthly_balance_prevQ']
balance_churn_no = data[data['churn'] == 0]['average_monthly_balance_prevQ']

# Prueba t para comparar los saldos promedio mensuales de ambos grupos
t_stat, p_val = ttest_ind(balance_churn_yes, balance_churn_no)
print(f'T-statistic: {t_stat}, P-value: {p_val}')
```

Como el p-value de la prueba t es menor a 0.05, podemos concluir que hay una diferencia estadísticamente significativa entre los saldos promedio mensuales del trimestre anterior de los clientes que abandonaron y los que no, lo que apoya nuestra hipótesis de que el saldo promedio mensual del trimestre anterior es una variable relevante.

**A.DETERMINAR EL MODELO A ELEGIR PARA EL DATASET**:

Se decide evaluar 3 tipos de modelos en el dataset : Regresión logística, Bosques aleatorios y Gradient Boosting con el fin de evaluar el que tenga un mayor rendimiento y precisión en datos.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# Preparación de datos del dataset
X = data.drop(columns=['churn', 'last_transaction'])
y = data['churn']
X = pd.get_dummies(X)  # Codificar variables categóricas
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Se definen los 3 modelos para identificar el de mayor certeza
models = {
    'Regresion logística': LogisticRegression(max_iter=1000),
    'Bosques aleatorios': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier()
}

# Por ende, se evalúan los modelos
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f'\n{name}')
    print('Matriz de Confusión:')
    print(confusion_matrix(y_test, y_pred))
    print('Reporte de Clasificación:')
    print(classification_report(y_test, y_pred))
    print('ROC AUC Score:')
    print(roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]))
```

**CONCLUSION** : En base a ello, se logra concluir que el modelo Random Forest y el Gradient Boosting tienen un desempeño similar en base a los resultados como ROC AUC SCORE (0,834 y 0,837 respectivamente) ,siendo ambos superior al de Regresion Logística en términos de capturar clientes churn (que es clase= 1). Siendo así, el modelo elegido el de Bosques aleatorios o Random Forest.
